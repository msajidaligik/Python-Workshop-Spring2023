## Color, Markings, and Line Styles

```python
data=np.random.random(50,).cumsum() 
```

```python
data
```

```python
plt.plot(data, color="black", marker='o', linestyle="dashed")
```

```python
%matplotlib inline
```

```python
plt.plot(data, color="blue", marker='.', linestyle="dashed")
```

```python
plt.plot(data, color='k', linestyle='dashed', drawstyle='steps-post')
```

```python
plt.plot(data, color='c', linestyle="dashed", drawstyle='steps-post', label='Stepwise')
plt.legend(loc='best')
```

## Subplots

```python
import random
random_floats = [[round(random.uniform(0, 10), 2) for _ in range(10)] for _ in range(4)]
```

```python
random_floats
```

```python
from matplotlib import pyplot as plt

def plot_data(X):
    plt.figure(1)
    k = 0
    plt.subplot2grid((2,2),(0,0))
    plt.plot(X[0])
    plt.xlabel("X value")
    plt.ylabel("Y value")
    plt.title("Subplot "+str(1))
    
    plt.subplot2grid((2,2),(0,1))
    plt.plot(X[0])
    plt.xlabel("X value")
    plt.ylabel("Y value")
    plt.title("Subplot "+str(2))
    
    plt.subplot2grid((2,2),(1,0))
    plt.plot(X[0])
    plt.xlabel("X value")
    plt.ylabel("Y value")
    plt.title("Subplot "+str(3))
    
    plt.subplot2grid((2,2),(1,1))
    plt.plot(X[0])
    plt.xlabel("X value")
    plt.ylabel("Y value")
    plt.title("Subplot "+str(4))
    
    plt.show()

plot_data(random_floats)
```

```python
from matplotlib import pyplot as plt

def plot_data(X):
    plt.figure(1)
    k = 0
    xx=np.arange(len(X[0]))
    plt.subplot2grid((2,2),(0,0))
    plt.bar(xx,X[0])
    plt.xlabel("X value")
    plt.ylabel("Y value")
    plt.title("Subplot "+str(1))
    
    plt.subplot2grid((2,2),(0,1))
    plt.scatter(xx,X[0])
    plt.xlabel("X value")
    plt.ylabel("Y value")
    plt.title("Subplot "+str(2))
    
    plt.subplot2grid((2,2),(1,0))
    plt.stem(X[0])
    plt.xlabel("X value")
    plt.ylabel("Y value")
    plt.title("Subplot "+str(3))
    
    plt.subplot2grid((2,2),(1,1))
    plt.plot(X[0])
    plt.xlabel("X value")
    plt.ylabel("Y value")
    plt.title("Subplot "+str(4))
    
    plt.show()

plot_data(random_floats)
```
