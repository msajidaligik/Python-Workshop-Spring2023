## Thickness, Label and Legend

```python
import random
random_floats = [[round(random.uniform(0, 10), 2) for _ in range(10)] for _ in range(4)]
```

```python
from matplotlib import pyplot

def plot_data(X):
    pyplot.figure(1)
    k = 0
    for i in range(0,2):
        for j in range(0,2):
            pyplot.subplot2grid((2,2),(i,j))
            pyplot.plot(X[k])
            pyplot.xlabel("X value")
            pyplot.ylabel("Y value")
            pyplot.title("Subplot "+str(k))
            k = k+1
    pyplot.show()

plot_data(random_floats)
```

```python
rng = np.random.RandomState(0)
```

```python
x = np.linspace(0, 10, 250)
```

```python
y = np.cumsum(rng.randn(250, 6), 0)
```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='best')
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```
